```python
final_data = new_data[['type','title','country','release_year','rating','listed_in','averageRating','show_id']]
```

```python
# final_data

final_data['country'] = final_data['country'].fillna('Other')
final_data['country'] = final_data.apply(lambda row: country(row), axis=1) 
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
final_data['clean_title'] = final_data['title'].apply(word_tokenize)
final_data['listed_in'] = final_data['listed_in'].apply(word_tokenize)
```

```python
list_stopwords = set(stopwords.words('english') + list(punctuation))
final_data['listed_in'] = final_data['listed_in'].apply(lambda x: [item for item in x if item not in list_stopwords])
final_data['clean_title'] = final_data['clean_title'].apply(lambda x: [item for item in x if item not in list_stopwords])
```

```python
genre_list = list(final_data["listed_in"].explode().unique())
def genre(genrelist):
    binaryList = []
    for genre in genre_list:
        if genre in genrelist:
            binaryList.append(1)
        else:
            binaryList.append(0) 
    return binaryList
final_data['listed_enc'] = final_data['listed_in'].apply(lambda x: genre(x))
final_data['listed_enc'].head()
```

```python
countrylist=[]
for i in final_data['country']:
    if i not in countrylist:
        countrylist.append(i)
def encoding_country(list_):
    binaryList = []  
    for coun in countrylist:
        if coun in list_:
            binaryList.append(1)
        else:
            binaryList.append(0)
    return binaryList
final_data['country_enc'] = final_data['country'].apply(lambda x: encoding_country(x))
final_data['country_enc'].head()
```

```python
type=list(final_data['type'].unique())
def encoding_type(list_):
    binaryList = []  
    for i in type:
        if i in list_:
            binaryList.append(1)
        else:
            binaryList.append(0)
    return binaryList
final_data['type_enc'] = final_data['type'].apply(lambda x: encoding_type(x))
final_data['type_enc'].head()
```

```python
# final_data[final_data['rating'].isnull()]
final_data.loc[final_data['rating'] == '74 min', 'rating'] ='R'
final_data.loc[final_data['rating'] == '84 min', 'rating'] ='R'
final_data.loc[final_data['rating'] == '66 min', 'rating'] ='R'
final_data.loc[final_data['show_id'] == 's5990', 'rating'] ='PG-13'
final_data.loc[final_data['show_id'] == 's7538', 'rating'] ='PG-13'
final_data.loc[final_data['show_id'] == 's6828', 'rating'] ='TV-MA'
final_data.loc[final_data['show_id'] == 's7313', 'rating'] ='TV-MA'
```

```python
mov_type=list(final_data['rating'].unique())
mov_type
def encoding_mov_type(list_):
    binaryList = []  
    for i in mov_type:
        if i in list_:
            binaryList.append(1)
        else:
            binaryList.append(0)
    return binaryList
final_data['rating_enc'] = final_data['rating'].apply(lambda x: encoding_mov_type(x))
final_data['rating_enc'].head()
```

```python
from scipy import spatial

def similarity(movieId1, movieId2):
    a = final_data.loc[final_data['show_id'] == movieId1]
    b = final_data.loc[final_data['show_id'] == movieId2]
    
    genresA = a['listed_enc'].values
    genresB = b['listed_enc'].values
    genreDistance = 1 - (spatial.distance.cosine(genresA[0], genresB[0]))
#     print(genreDistance)
    
    counA = a['country_enc'].values
    counB = b['country_enc'].values
    counDistance = 1 - (spatial.distance.cosine(counA[0], counB[0]))
#     print(genreDistance)
    
    typeA = a['type_enc'].values
    typeB = b['type_enc'].values
    typeDistance = 1 - (spatial.distance.cosine(typeA[0], typeB[0]))
    
    audience_typeA = a['rating_enc'].values
    audience_typeB = b['rating_enc'].values
    audience_typeDistance = 1 - (spatial.distance.cosine(audience_typeA[0], audience_typeB[0]))
    
    ratingA = a['averageRating'].values
    ratingB = b['averageRating'].values
    ratingDistance = 1 - (abs(ratingA[0]-ratingB[0])/8.8)
    
    yearA = a['release_year'].values
    yearB = b['release_year'].values
#     print(yearA[0])
#     print(yearB[0])
    yearDistance = 1- (abs(yearA[0]-yearB[0])/96)
#     print(yearDistance)
    
    return genreDistance + counDistance  + typeDistance + audience_typeDistance + ratingDistance + yearDistance
```
